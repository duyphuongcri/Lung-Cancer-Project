## Overall Stage Analysis

```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 4
Ensemble = (567, 479, 351, 287)
Clinical = (649, 494, 362, 285)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, Ensemble, bar_width,
alpha=opacity,
color='b',
label='Ensemble')

rects2 = plt.bar(index + bar_width, Clinical, bar_width,
alpha=opacity,
color='g',
label='Clinical')

plt.xlabel('Overall Stage')
plt.ylabel('MAE (days)')
plt.title('Overall Stage Analysis')
plt.xticks(index + bar_width, ('Stage 1', 'Stage 2', 'Stage 3', 'Stage 4'))
plt.legend()

plt.tight_layout()
plt.show()
```
